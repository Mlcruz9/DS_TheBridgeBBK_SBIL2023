A veces encontraremos tablas en páginas web de las que no nos permiten un enlace de descarga como tal. Vamos a ver un sencillo ejemplo práctico de cómo extraerlas

```python
# Leemos la información de la página

tabla_poblacion = pd.read_html("https://www.worldometers.info/world-population/population-by-country/")

print(tabla_poblacion)
```

```python
tabla_poblacion[0].info()
```

```python
tabla_poblacion[0].to_csv("data/poblacion-mundial_2020.csv", index = False)
```

```python
# Ejemplo 2

tablas_ventas = pd.read_html("https://es.wikipedia.org/wiki/Anexo:%C3%81lbumes_musicales_m%C3%A1s_vendidos")

len(tablas_ventas)
```

```python
tablas_ventas[5].info()
```

```python
tablas_ventas[1:] # estas 5 tablas tienen la información que nos interesa
```

```python
numero_tabla = 1
for tabla in tablas_ventas[1:]:
    tabla.to_csv(f"data/ventas_tabla-{numero_tabla}.csv", index=False)
    numero_tabla += 1
```
