### Step 4. Which continent drinks more beer on average?

```python
beer_grouped = drinks[['beer_servings', 'continent']].groupby('continent').mean().sort_values(by= 'beer_servings', ascending= False)
```

```python
beer_grouped
```

```python
beer_grouped[beer_grouped['beer_servings'] == beer_grouped['beer_servings'].max()].index[0]
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks[['continent', 'wine_servings']].groupby('continent').describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks[['continent', 'total_litres_of_pure_alcohol']].groupby('continent').mean()
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks[['beer_servings', 'continent', 'wine_servings']].groupby('continent').mean(numeric_only=True)
```

### Step 8. Print the mean, min and max values for spirit consumption per continent.

```python
drinks[['spirit_servings', 'continent']].groupby('continent').aggregate(['mean', 'min', 'max'])
```
