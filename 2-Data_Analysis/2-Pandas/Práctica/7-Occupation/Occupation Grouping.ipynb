# Occupation

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep= '|')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users[['occupation', 'age']].groupby('occupation').mean().round()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users['gender_num'] = np.where(users['gender'] == 'M', 1, 0)
```

```python
#(users.groupby('occupation')["gender_num"].sum()/users.groupby('occupation')["gender"].count()).round(2).sort_values(ascending = False) -- Con la media usando count y sum
```

```python
users[['occupation', 'gender_num']].groupby(['occupation']).mean().sort_values(by='gender_num', ascending= False)
```

```python
users[['gender', 'occupation']].groupby('occupation').value_counts(normalize=True).unstack()\
    [['M']].sort_values(by='M', ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users[['occupation', 'age']].groupby('occupation').agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users[['occupation', 'gender', 'age']].groupby(['gender', 'occupation'])[['age']].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
pd.DataFrame(users[['occupation', 'gender']].groupby('occupation').value_counts(normalize= True)*100)
```

```python
users[['gender_num', 'occupation']].groupby('occupation').mean()
```

```python
users.groupby(['occupation', 'gender'])['age'].count()
```

```python
users.groupby('occupation')['age'].count()
```
