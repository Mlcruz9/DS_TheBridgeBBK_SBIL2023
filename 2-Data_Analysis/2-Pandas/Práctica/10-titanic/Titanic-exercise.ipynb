34. Using `kdeplot` from `seaborn` represent those who not survived distributed by age.

Hint: First you should "filter" the `titanic` dataset where the column "survived" is 0, indexing the column `"age"` only.

Arguments you should pass to the function: 

    - color = "red"
    - label = "Not Survived"
    - shade = True
    
You should get something like this: 

![kdeplot](./img/kdeplotsur.png)

```python
titanic_deaths_2 = titanic['age'][titanic['survived'] == 0]
titanic_deaths_3 = titanic['age'][titanic['survived'] == 1]
```

```python
sns.kdeplot(data = titanic_deaths_2, color= 'red', label = 'not_survived', shade = True)
sns.kdeplot(data = titanic_deaths_3, color= 'green', label = 'survived', shade = True)

plt.legend();
```
