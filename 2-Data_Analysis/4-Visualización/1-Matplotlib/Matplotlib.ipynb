# Básicos de matplotlib

**Matplotlib** es una biblioteca de visualización de datos multiplataforma construida sobre arreglos de **NumPy**, y diseñada para trabajar con el amplio conjunto de herramientas de SciPy.

Permite crear gráficos estáticos, interactivos y de alta calidad. Fue creada con el objetivo de **facilitar la representación gráfica de datos** y resultados en un formato comprensible y estéticamente agradable. Con Matplotlib, los usuarios pueden crear una amplia variedad de gráficos, incluyendo **gráficos de líneas**, **gráficos de dispersión**, **histogramas**, **gráficos de barras**, **gráficos de torta**, **gráficos de contorno**, **diagramas de caja** y más.

Las principales características de Matplotlib incluyen su flexibilidad y capacidad para trabajar con diferentes sistemas operativos, así como su integración con NumPy, una biblioteca fundamental para la computación numérica en Python. Matplotlib ofrece una amplia gama de opciones de personalización para controlar el aspecto y el estilo de los gráficos, lo que permite adaptarlos según las necesidades específicas del usuario o el proyecto.

Esta biblioteca es ampliamente utilizada en el ámbito de la **ciencia de datos**, la investigación, la ingeniería y cualquier otra disciplina que requiera la visualización de datos para análisis, **presentación o comunicación efectiva de resultados**. Además, Matplotlib sirve como base para otras bibliotecas de visualización más especializadas y de alto nivel, como **Seaborn y Pandas**, que simplifican la creación de gráficos complejos con menos líneas de código y una sintaxis más sencilla.

## *Lineplots*

Un line plot en Matplotlib es un tipo común de gráfico utilizado para representar datos en un espacio bidimensional. Este tipo de gráfico es especialmente útil para mostrar la relación entre dos variables continuas a lo largo de un eje.

````python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-') #Primer trazo (Continuo)
plt.plot(x, np.cos(x), '--'); #Segundo trazo (Discontinuo)
# plt.plot(x, x**2, '.');
````

Notar que creamos una variable "fig", que representaría un objeto vacío en el que luego vamos creando trazos. Este es uno de los enfoques o maneras en las que podemos realizar plots con matplotlib. Podemos salvar posteriormente la imagen con *".savefig"*.

````python
fig.savefig('img/my_figure.png')
````

### Enfoques en matplotlib

Podemos crear imagenes de matplotlib de dos maneras (con dos sintaxis) distintas, dependiendo de cómo queramos interactuar con el objeto. La primera declarar el objeto sin asignarlo a variables, la segunda asignandolo a una variable, a la que luego podemos asignar cortes que podremos modificar posteriormente (plotear elementos sobre los cortes).

A continuación el primer enfoque:

````python
plt.figure()#Se  inicializa la figura

plt.subplot(2, 1, 1) # filas, columnas, posición
plt.plot(x, np.sin(x))


plt.subplot(2, 1, 2) # filas, columnas, posición
plt.plot(x, np.cos(x), c= 'r')
plt.plot(x, np.sin(x), '--', c= 'g');
````

Nótese que en el enfoque anterior no se declaran variables, vamos con el segundo enfoque:

````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2, 2)
ax[0, 1].plot(x, np.sin(x))

ax[1, 0].plot(x, np.cos(x));
````

### Acciones que podemos realizar en los lineplots

Matplotlib es bastante modificable y podemos realizar múltiples acciones en los diferentes gráficos, como **cambiar colores de plots**, **cambiar estilo de linea**, **ajustar límites de ejes**, **nombrar los ejes**, **crear leyendas** y más.

````python
plt.style.use('seaborn-whitegrid')
````

#### Colores de lineas

A continuación cambiamos los colores de las lineas, si no se especifica un color matplotlib lo agrega por defecto.

````python
plt.plot(x, np.sin(x - 0), color='blue')        # Color por nombre
plt.plot(x, np.sin(x - 1), color='g')           # Nombre corto (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Escala de grises entre 0 y 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Código hexagesimal (RRGGBB de 00 a FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # Tupla RGB, de 0 a 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # Colores HTML
````

#### Estilo de linea

Podemos a su vez cambiar el estilo de linea.

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# Código corto:
plt.plot(x, x + 4, linestyle='-')  # Solida
plt.plot(x, x + 5, linestyle='--') # Cortada
plt.plot(x, x + 6, linestyle='-.') # Corte y punto
plt.plot(x, x + 7, linestyle=':');  # Puntos
````

Tabién podemos, por qué no, cambiar estilo de linea y especificar color a la vez.

````python
plt.plot(x, x + 0, '-g')  # Verde solido
plt.plot(x, x + 1, '--c') # Celeste cortado
plt.plot(x, x + 2, '-.k') # Negro corte y punto
plt.plot(x, x + 3, ':r');  # Rojo puntos
````

#### Agregar una legenda

Acá un ejemplo de agregar una leyenda:

````python
plt.plot(x, np.sin(x), '-g', label = 'Seno sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'Coseno cos(x)')
plt.axis('equal')
plt.grid(False)


plt.legend(loc = 'lower right', frameon = True, fancybox=True, framealpha=0.7, shadow=True, borderpad=0.5);
````

Un ejemplo más potente:

````python
cities = pd.read_csv('data/california_cities.csv')

# Data que queremos del .csv
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

plt.figure(figsize=(10,5))
# Sin etiqueta, colores c, colormap, tamaño s, ancho de linea, transparencia
plt.scatter(lon, lat, label=None,
            c=np.log10(population), cmap='viridis',
            s=area, linewidth=0, alpha=0.5)

plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

# Creamos leyenda:
# Areas de distintos tamañps
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area,
                label=str(area) + ' km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Area')

plt.title('California Cities: Area and Population');
````

## *Scatterplots*

Podemos hacer scatterplots de manera muy similar a los lineplots.

````python
x = np.linspace(0, 10, 20)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
````

O con ``plt.scatter()``

````python
plt.scatter(x, y);
````

````python
rng = np.random.RandomState(0)
x = rng.randn(100) # edades
y = rng.randn(100) # alturas
colors = rng.rand(100) # pais
sizes = 1000 * rng.rand(100) # numero personas

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='plasma');
plt.colorbar();
````

## Matplotlib Gotchas

La matoria de funciones de ``plt`` se utilizan de la misma manera en los dos enfoques antes mencionados (como ``plt.plot()`` → ``ax.plot()``, ``plt.legend()`` → ``ax.legend()``, etc.), pero algunos tienen cambios.
Funciones para colocar limites, etiquetas, y títulos pueden ser un poco distintos.
**Algunos ejemplos:**

- ``plt.xlabel()``  → ``ax.set_xlabel()``
- ``plt.ylabel()`` → ``ax.set_ylabel()``
- ``plt.xlim()``  → ``ax.set_xlim()``
- ``plt.ylim()`` → ``ax.set_ylim()``
- ``plt.title()`` → ``ax.set_title()``

Con ``ax.set()`` podemos hacer varios cambios a la vez:

```python

ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot'):
```
